```python
## Objects and classes

class Insurance:
    def __init__(self, type, coverage, cust_name, age, gender):
        self.type = type
        self.coverage = coverage
        self.cust_name = cust_name
        self.age = age
        self.gender = gender

    ## utilizing instance methods
    def patient_details(self):
        print(f"1. Name of the individual is: {self.cust_name}\n"
              f"2. The type of the insurance is: {self.type}\n"
              f"3. The coverage of the plan is: {self.coverage}\n"
              f"4. The age of the individual is: {self.age}\n"
              f"5. The gender of the person is: {self.gender}")

# Create an instance
TataAIG = Insurance("life", 100000, "Raj Patel", 25, "Male")

# Call the method
TataAIG.patient_details()
```

```python
## Modelling a Bank Account

class Account:
    def __init__(self, name, balance=0):
        self.name = name
        self.balance = balance
        self.initial_balance = balance
        self.last_withdrawn = 0  # to store last withdrawn amount

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited {amount}. New balance is {self.balance}.")
        return self.balance

    def withdraw(self, amount):
        if amount > self.balance:
            print(f"Withdrawal of {amount} failed. Available balance: {self.balance}.")
        else:
            self.balance -= amount
            self.last_withdrawn = amount  # update last withdrawn amount
            print(f"Withdrew {amount}. Remaining balance: {self.balance}.")
        return self.balance

    def updated_balance(self):
        print(f"1. Name of the person: {self.name}")
        print(f"2. Initial balance: {self.initial_balance}")
        print(f"3. Last withdrawn amount: {self.last_withdrawn}")
        print(f"4. Updated balance: {self.balance}")
        return self.balance

# Example usage
person1 = Account("Rajesh", 5000)
person1.deposit(1000)
person1.withdraw(1500)
person1.updated_balance()
```

```python
## Inheritance is the concept of the object oriented programming where we inherit attribute and the methods from the another class. 
# It consist of single inheritance and multiple inheritance.
class Account():
    def __init__(self,account_holder_name,account_number,balance,account_type):
        self.account_holder_name = account_holder_name
        self.account_number = account_number
        self.balance = balance
        self.account_type = account_type

    def deposit(self,amount):
        self.balance +=amount

    def withdraw(self,amount):
        self.balance -=amount

    def display_balance(self):
        print(f"Updated Account Balance is: {self.balance}") 
    
    def display_account_info(self):
        print(f"The account holder name is:{self.account_holder_name}\n"
              f"The account holder account number is: {self.account_number}\n"
              f"The account balance is:{self.balance}")

class SavingsAccount(Account):
    def __init__(self,account_holder_name,account_number,balance,account_type,interest_rate):
        super().__init__(account_holder_name,account_number,balance,account_type)
        self.interest_rate = interest_rate

    def apply_interest(self):
        self.balance = self.balance+(0.075*self.balance)
        print(f"The applied interest rate is 7.5% {self.balance}")

    def minimum_balance_required(self):
        if self.balance>5000:
            print("The account has sufficient balance!!")
        else:
            print("The account has insufficeint balance!!")

    
class CheckingAccount(Account):
            def __init__(self,account_holder_name,account_number,balance,account_type,overdraft_limit):
                super().__init__(account_holder_name,account_number,balance,account_type)
                self.overdraft_limit = overdraft_limit
                
            def withdraw(self,amount):
                if amount>self.balance +self.overdraft_limit:
                      print("Withdrawal failed. Overdraft limit exceeded.")
                else:
                    self.balance -= amount
                    print(f"Withdrawn: {amount}. New balance: {self.balance}")

```

```python
## Encapsulation and Abstraction
##  Encapsulation with Getter and Setter Methods
## Public, Private and Protected
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance  # private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient funds!")

    def get_balance(self):
        return self.__balance
```

```python
## Polymorphism

class Animal:
    def speak(self):
        print("The animal makes a sound")

class Dog(Animal):
    def speak(self):
        print("The dog barks")

class Cat(Animal):
    def speak(self):
        print("The cat meows")

def make_animal_speak(animal):
    animal.speak()

make_animal_speak(Dog())  
make_animal_speak(Cat())  
```

```python
from abc import ABC, abstractmethod

#  Abstract Base Class (Abstraction)
class PaymentMethod(ABC):
    def __init__(self, payer_name):
        self.payer_name = payer_name

    @abstractmethod
    def authorize_payment(self, amount):
        pass

    @abstractmethod
    def process_payment(self, amount):
        pass

#  Credit Card Payment (Encapsulation + Polymorphism)
class CreditCardPayment(PaymentMethod):
    def __init__(self, payer_name, card_number, cvv, expiry_date):
        super().__init__(payer_name)
        self.__card_number = card_number
        self.__cvv = cvv
        self.expiry_date = expiry_date

    def authorize_payment(self, amount):
        print(f"Authorizing credit card payment for {self.payer_name}")
        return True  # Simulated

    def process_payment(self, amount):
        if self.authorize_payment(amount):
            print(f"Processed credit card payment of ${amount} for {self.payer_name}")

#  PayPal Payment (Encapsulation + Polymorphism)
class PayPalPayment(PaymentMethod):
    def __init__(self, payer_name, email, password):
        super().__init__(payer_name)
        self.__email = email
        self.__password = password

    def authorize_payment(self, amount):
        print(f"Authorizing PayPal payment for {self.payer_name}")
        return True  # Simulated

    def process_payment(self, amount):
        if self.authorize_payment(amount):
            print(f"Processed PayPal payment of ${amount} for {self.payer_name}")

# UPI Payment (Encapsulation + Polymorphism)
class UPIPayment(PaymentMethod):
    def __init__(self, payer_name, upi_id, pin):
        super().__init__(payer_name)
        self.__upi_id = upi_id
        self.__pin = pin

    def authorize_payment(self, amount):
        print(f"Authorizing UPI payment for {self.payer_name}")
        return True  # Simulated

    def process_payment(self, amount):
        if self.authorize_payment(amount):
            print(f"Processed UPI payment of ${amount} for {self.payer_name}")
```

```python
def run_payments():
    credit = CreditCardPayment("Rajesh", "1234-5678-9999", "123", "12/26")
    paypal = PayPalPayment("Seema", "seema@example.com", "securepass")
    upi = UPIPayment("Karan", "karan@upi", "0000")

    # All behave differently but use the same interface
    credit.process_payment(2500)
    paypal.process_payment(1800)
    upi.process_payment(999)

run_payments()
```
